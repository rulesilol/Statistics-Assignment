```julia
using Random
using LinearAlgebra
using Plots
using Distributions
using StatsBase
```

## For N = 50

```julia
N = 50
Wigner = zeros(N, N)
for i in 1:N
    for j in 1:N
        if i <= j
            Wigner[i,j] = randn() + 1
        end
    end
end
for i in 1:N
    for j in 1:N
        if i>j
            Wigner[i, j] = Wigner[j, i]
        end
    end
end

evals = eigvals(Wigner)/sqrt(N)
evals = evals[1:N-1];
```

```julia
empirical = ecdf(evals)
x_grid = -2:0.01:2
plot(x_grid,empirical.(x_grid))
```

```julia
h = fit(Histogram, evals, nbins = 10)
h = normalize(h)
plot(h, title = "N = 50")
f(x) = sqrt(4-(x^2))/(2*pi)
plot!(f, -2, 2)
```
